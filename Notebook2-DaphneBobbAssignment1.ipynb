It might make it easier for buyers to know what part of Ireland to go to for used cars. First, we'll find the frequency of each town in the Location column: 

```python
frequencies = carSales_df[" Location"].value_counts()
frequencies
```

Let's write a cool little method to allow the user to see which locations have the most cars of their specified price range. This will then be visualized as a bar chart based on the frequencies and locations. Since the frequency table we get is a Pandas series, to get our x and y values for the bar chart, we can use the values and index functions to extract the locations and the number of cars from our frequency table. 

```python
import matplotlib 
import matplotlib.pyplot as plt 
```

```python
def locationForPrice(price): 
    userCriteria = carSales_df.loc[(carSales_df[" Sale Price"] < price)]
    frequencyTable = userCriteria[" Location"].value_counts()

    y_pos = list(range(len(frequencyTable.index)))
    y_pos

    plt.figure(figsize = (13, 3))
    plt.bar(frequencyTable.index, frequencyTable.values, align="center", color="coral")
    plt.ylabel("# of Cars")
    plt.title("Cars sold in budget by location")

    return plt.show()
    
```

Now let's put our method to work with a few examples: 

```python
locationForPrice(5000)
```

```python
locationForPrice(10000)
```

```python
locationForPrice(50000)
```

```python
locationForPrice(100000)
```
